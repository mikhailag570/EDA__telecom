```python
# Можно использовать словери в качестве аргументов

d = {"No": False, "Yes": True}
df["International plan"] = df["International plan"].map(d)
df.head()
```

```python
# Также можно использовать replace

df = df.replace({"Voice mail plan": d})
df.head()
```

```python
# Группировка в pandas

columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

```python
# Или передадим в agg список функций.

columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
# Построим сводную таблицу Churn и International plan

pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

```python
# Посмотрим на среднее количество дневных, вечерних и ночных звонков по кодам городов

df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```

```python
# Если нужно общее количество звонков для всех пользователей, создадим серию total_calls и вставим ее в DataFrame.

total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)

df.head()
```

```python
# Или проще, не создавая промежуточный экземпляр Series:

df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

```python
# Чтобы удалить столбцы или строки, используем метод drop, передав необходимые индексы 
# и параметр оси (1, если вы удаляете столбцы, и ничего, или 0, если вы удаляете строки).

df.drop(["Total charge", "Total calls"], axis=1, inplace=True)

# или от строк 1 и 2
df.drop([1, 2]).head()
```

```python
'''
Попробуем предсказать отток клиентов и как он связан с международной тарификацией.
'''
```

```python
pd.crosstab(df["Churn"], df["International plan"], margins=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# В формате Retina более четкая и разборчивая графика.
%config InlineBackend.figure_format = 'retina'
```

```python
sns.countplot(x="International plan", hue="Churn", data=df)
```

```python
'''
Мы видим, что с International Plan уровень оттока намного выше. 
Возможно, большие и плохо контролируемые расходы на международные звонки очень конфликтны 
и приводят к недовольству клиентов оператора связи.
'''
```

```python
# Рассмотрим еще одну важную функцию – звонки в службу поддержки клиентов.

pd.crosstab(df["Churn"], df["Customer service calls"], margins=True)
```

```python
sns.countplot(x="Customer service calls", hue="Churn", data=df)
```

```python
'''
Увеличивается отток начиная с 4-го звонка
'''
```

```python
# Теперь давайте добавим в наш DataFrame бинарную функцию — Звонки в службу поддержки > 3. 
# И еще раз посмотрим, как она связана с оттоком клиентов.

df["Many_service_calls"] = (df["Customer service calls"] > 3).astype("int")

pd.crosstab(df["Many_service_calls"], df["Churn"], margins=True)
```

```python
sns.countplot(x="Many_service_calls", hue="Churn", data=df)
```

```python
'''
Построим еще одну таблицу непредвиденных обстоятельств, которая связывает отток как с международным планом, 
так и с недавно созданными Many_service_calls.
'''
```

```python
pd.crosstab(df["Many_service_calls"] & df["International plan"], df["Churn"])
```

```python
'''
Таким образом, прогнозируя нелояльность клиента (Отток=1) в случае, когда количество обращений в сервисный центр больше 3 
и добавлен Международный план (и прогнозируя Отток=0 в противном случае), 
можно ожидать точности 85,8%.
Это число, 85,8%, которое мы получили в результате этих очень простых рассуждений, 
'''
```
